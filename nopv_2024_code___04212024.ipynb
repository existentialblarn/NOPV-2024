```python
# remove columns
columns_to_drop = ['Unnamed: 3', 'Unnamed: 4']  # List of column names to drop
df = df.drop(columns_to_drop, axis=1)
df.head()
```

```python
# rename columns

df.rename(columns={'THE CITY OF NEW YORK - DEPARTMENT OF FINANCE': 'boro',
                    'Unnamed: 1': 'block',
                    'Unnamed: 2': 'lot',
                    'Unnamed: 5': 'add_1',
                    'Unnamed: 6': 'add_2',
                    'Unnamed: 7': 'cit_sta_zip',
                    'Unnamed: 8': 'country',
                    'Unnamed: 9': 'tax_cl',
                    'Unnamed: 10': 'bldg_cl',
                    'Unnamed: 11': 'og_mark_val',
                    'Unnamed: 12': 'og_assd_val',
                    'Unnamed: 13': 'og_exempt',
                    'Unnamed: 14': 'og_trans_assd_val',
                    'Unnamed: 15': 'og_trans_exempt',
                    'Unnamed: 16': 'og_tax_val',
                    'Unnamed: 17': 'rev_mark_val',
                    'Unnamed: 18': 'rev_assd_val',
                    'Unnamed: 19': 'rev_exempt',
                    'Unnamed: 20': 'rev_trans_assd_val',
                    'Unnamed: 21': 'rev_trans_exempt',
                    'Unnamed: 22': 'rev_tax_val',
                    'Unnamed: 23': 'rc1',
                    'Unnamed: 24': 'rc2',
                    'Unnamed: 25': 'rc3',
                    'Unnamed: 26': 'rc4',
                    'Unnamed: 27': 'rc5'
                    
                    }, inplace=True)
df.head()
```

```python
# concat BBL [(boro cell* 1000000000) + (block cell* 10000) + lot cell]

def bbl_formula(row):
    return (row['boro'] * 1000000000) + (row['block'] * 10000) + (row['lot'])

# apply BBL formula to create a new column
df['BBL'] = df.apply(bbl_formula, axis=1) # .apply() method is used to apply a function along an axis of a dataframe or series

# specify BBL position
position = 3

#insert BBL concat at the position
df.insert(position, 'BBL', df.pop('BBL')) # .pop first removes the 'BBL' column from its current position and then inserts it at the specified position.

df.head()
```

```python
# Add File date

date_to_add = '2024-04-04'

# Convert the date to a pandas Timestamp object
date_object = pd.to_datetime(date_to_add)

# Add the date to all rows
df['nopv_file_date'] = date_object

# specify File date position
position = 27

#insert BBL concat at the position
df.insert(position, 'nopv_file_date', df.pop('nopv_file_date'))

df.head()
```

```python
# Add Column: Change in MV

def change_in_mv(row):
    return row['rev_mark_val'] - row ['og_mark_val']

# apply formula to create a new column
df['change_in_mv'] = df.apply(change_in_mv, axis=1)

# specify change in mv position
position = 27

#insert change in mv at the position
df.insert(position, 'change_in_mv', df.pop('change_in_mv'))

df.head()
```

```python
# Add Column: Change in Taxable Value

def change_in_tax_val(row):
    return row['rev_tax_val'] - row ['og_tax_val']

# appl BBL formula to create a new column
df['change_in_tax_val'] = df.apply(change_in_tax_val, axis=1)

# specify BBL position
position = 28

#insert BBL concat at the position
df.insert(position, 'change_in_tax_val', df.pop('change_in_tax_val'))

df.head()
```

```python
# Rename reason 1?

def reason_condition(row):
    if row['rc1'] == 'A':
        return 'Alteration'
    elif row['rc1'] == 'AP':
        return 'Apportionment'
    elif row['rc1'] == 'B':
        return 'Building in Progress Last Year'
    elif row['rc1'] == 'D':
        return 'Demolition'
    elif row['rc1'] == 'E':
        return 'Full Exempt and is now restored to taxable status'
    elif row['rc1'] == 'E0':
        return 'Sales'
    elif row['rc1'] == 'E1':
        return 'Economics'
    elif row['rc1'] == 'E2':
        return 'Administrative Review (Assessor initiated)'
    elif row['rc1'] == 'E3':
        return 'Sales (Taxpayer initiated Change)'
    elif row['rc1'] == 'E4':
        return 'Economics (Taxpayer initiated Change)'
    elif row['rc1'] == 'E5':
        return 'Administrative Review (Taxpayer initiated)'
    elif row['rc1'] == 'M':
        return 'MV only Changes'
    elif row['rc1'] == 'N':
        return 'New Building'
    elif row['rc1'] == 'T':
        return 'Transfer to or from REUC'
    elif row['rc1'] == 'NP':
        return 'New Building in Progress'
    elif row['rc1'] == 'P':
        return 'Physical Change'
    elif row['rc1'] == 'S':
        return 'Gain or Loss to Street'
    elif row['rc1'] == 'X':
        return 'Change other than fully exempt restored'
    else:
        return ''

df['rc1_named'] = df.apply(reason_condition, axis=1)

# specify renamed reason position
position = 29

#insert BBL concat at the position
df.insert(position, 'rc1_named', df.pop('rc1_named'))

df.head()
```

```python
# Merge with prior NOPV
df1 = df
df2 = pd.read_excel(r"C:\Users\AbarentosG\RPT\NOPV_2\02_clean-data\NOPV-02162024_03142024.xlsx") #rename with latest prior data

# Sort both df by the timestamp column in descending order
df1 = df1.sort_values(by='nopv_file_date', ascending=False)
df2 = df2.sort_values(by='nopv_file_date', ascending=False)

# Concatenate the dfs while ignoring the index
merged_df = pd.concat([df1, df2], ignore_index=True)

print((merged_df.head()), (merged_df.shape))
```

```python
# Remove duplicates based on the 'BBL' column, keeping only the first occurrence
merged_df = merged_df.drop_duplicates(subset='BBL', keep='first')

print((merged_df.head()), (merged_df.shape))
```

```python
#filter class 1
desired_cat = ['1A','1B','1C','1D',"1"]
filtered_c1 = merged_df[merged_df['tax_cl'].astype(str).isin(desired_cat)]

#set display options
pd.set_option('display.float_format', '{:,.2f}'.format)

# Generate summary table with counts for reason category
sum_tab_count_1 = filtered_c1['rc1_named'].value_counts().reset_index()
sum_tab_count_1.columns = ['Reason', '# of Parcels']

# Calculate the sum of the 'Change in Taxable Value' column for each category
sums_1 = filtered_c1.groupby('rc1_named')['change_in_tax_val'].sum().reset_index()
sums_1.columns = ['Reason', 'Change in Taxable AV']

# Calculate Total Tax *.12283
tax_tot_1 = sums_1['Total Tax'] = sums_1['Change in Taxable AV'] * 0.12283

#add grand total
grand_tot_count_1 = sum_tab_count_1['# of Parcels'].sum()

# Append a row for the grand total to the # parcels df
grand_tot_row_count_1 = pd.DataFrame({'Reason': ['Grand Total'], '# of Parcels': [grand_tot_count_1]})
sum_tab_count_1 = pd.concat([sum_tab_count_1, grand_tot_row_count_1], ignore_index=True)

# Append a row for the grand total to the Change in Tax AV df
grand_tot_sum_1 = sums_1['Change in Taxable AV'].sum()
grand__sum_1 = sums_1['Total Tax'].sum()
grand_tot_row_sum_1 = pd.DataFrame({'Reason': ['Grand Total'], 'Change in Taxable AV': [grand_tot_sum_1], 'Total Tax': [grand__sum_1]})
sums_1 = pd.concat([sums_1, grand_tot_row_sum_1], ignore_index=True)


# Merge the counts and sums into a single summary table
sum_tab_1 = pd.merge(sum_tab_count_1, sums_1, on='Reason')

print(sum_tab_1)
```
